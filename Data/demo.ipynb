```python
import pandas as pd
import numpy as np
import json 
import random
```

```python
armoni = pd.read_excel("ÇANAKKALE ARMONİ BAZ LİSTE.xlsx")

sah = list()

for i in armoni["MEVKİ"]:
    sah.append(i[:-1])

armoni["MEVKİ"] = sah

armoni["FİYAT"] = [3703*i for i in armoni["M2 NET"]]

armoni.head()
```

```python
ayvo = pd.read_excel("AYVO BAZ LİSTE.xlsx")
ayvo["FİYAT"] = [3703*i for i in ayvo["M2 NET"]]

ayvo.head()
```

```python
izmit = pd.read_excel("İZMİT BAZ LİSTE.xlsx")
izmit["FİYAT"] = [5250*i for i in izmit["M2 NET"]]

izmit.head()
```

```python
iznik = pd.read_excel("İznik Baz Liste Fiyatlandırılmış.xlsx")
iznik["FİYAT"] = [6500*i for i in iznik["M2 NET"]]

iznik.head()
```

```python
gumus = pd.read_excel("Gümüşova Baz Liste.xlsx")
gumus["FİYAT"] = [4500*i for i in gumus["M2 NET"]]

gumus.head()
# gumus.dropna(inplace=True)
```

```python
columns = ["İL", "İLÇE", "MEVKİ", "ADA", "PARSEL", "M2 NET", "FİYAT"]

data = armoni[columns]


data = pd.concat([data, ayvo[columns]])
data = pd.concat([data, izmit[columns]])
data = pd.concat([data, iznik[columns]])
data = pd.concat([data, gumus[columns]])



data
```

```python
set(data["MEVKİ"].values)
```

```python
import json
import random

data_list = []

for i in range(1, data.index.size + 1):
    parcel = {
        "model": "app.parcel",
        "pk": i,
        "fields": {
            "il": data["İL"].values[i-1],
            "ilce": data["İLÇE"].values[i-1],
            "mevki": data["MEVKİ"].values[i-1],
            "ada": data["ADA"].values[i-1],
            "parsel": str(data["PARSEL"].values[i-1]),
            "m2_net" : int(data["M2 NET"].values[i-1]),
            "fiyat" : int(data["FİYAT"].values[i-1])
        }
    }
    data_list.append(parcel)

with open("data.json", "w", encoding="utf-8") as outfile:
    json.dump(data_list, outfile, ensure_ascii=False)
```
