```python
import pandas as pd
import numpy as np
import json 
import random
```

```python
df = pd.read_excel("parsel_demo copy.xlsx")
df.fillna(0, inplace=True)
df
```

```python
df['SATIŞ DURUMU'] = df['SATIŞ DURUMU'].str.lower()




df
```

```python
set(list(df["SATIŞ DURUMU"]))
```

```python
satis_list = list()

for i in df["SATIŞ DURUMU"]:

    if i == "kapalı" or i == "satıldı":
        satis_list.append("onaylandi")
    else:
        satis_list.append("uygun")

df["SATIŞ DURUMU"] = satis_list

set(satis_list)

```

```python
import json
import random

data_list = []
data = df
for i in range(1, data.index.size + 1):
    parcel = {
        "model": "app.parcel",
        "pk": i,
        "fields": {
            "il": data["İL"].values[i-1],
            "ilce": data["İLÇE"].values[i-1],
            "mevki": data["MEVKİ"].values[i-1],
            "ada": data["ADA"].values[i-1],
            "parsel": str(data["PARSEL"].values[i-1]),
            "m2_net" : int(data["M2 NET"].values[i-1]),
            "fiyat" : int(data["FİYAT"].values[i-1]),
            "durum": data["SATIŞ DURUMU"].values[i-1],
            "menu_name": data["MENUNAME"].values[i-1]
        }
    }
    data_list.append(parcel)

with open("data_all_parcel.json", "w", encoding="utf-8") as outfile:
    json.dump(data_list, outfile, ensure_ascii=False)
```
